#Creating Artificial Neural Network using Sequential model from Keras/tf

#Generating data


```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

**Example data**:


*   An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial.
*   The trial had 2100 participants. Half were under 65 years old, half were 65 or older
* Around 95% of patients 65 or older had side effects
* Around 95% of patients under 65 did'nt have side effects

*Process of creating our dataset*

```python
for i in range(50):
  #The 5% of younger individuals who did experience side effects
  random_younger = randint(13, 64)
  train_samples.append(random_younger)
  train_labels.append(1)

  #The 5% of older individuals who did not experience side effects

  random_older = randint(65, 100)
  train_samples.append(random_older)
  train_labels.append(0)

for i in range(1000):

  #The 95%% of younger individuals who did not experience side effects

  random_younger = randint(13, 64)
  train_samples.append(random_younger)
  train_labels.append(0)

  #The 95%% of older individuals who did experience side effects


  random_older = randint(65, 100)
  train_samples.append(random_older)
  train_labels.append(1)
```

```python
for i in train_samples:
 print(i)
```

```python
for i in train_labels:
  print(i)
```

*Now data is generated, we're going to process it.*

*Next step is to pass these "Numpy arrays" to fit function*

#Processing to be a NumPy array format


```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

#Rescaling data to be between 0 and 1

```python
scaler = MinMaxScaler(feature_range=(0,1)) #Here we're transforming 13-100 into 0s and 1s
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1)) #This reshape is just formality cuz it function doesn't accept one D data by Default
```

*Now we are going to see integers have been transformed in some values between 0 and 1*

```python
for i in scaled_train_samples:
  print(i)
```

* Raw data was generated
* Processed to be NumPyArray format our model expects
* Rescaled data to be on a scale between 0 and 1


#Creating Artificial Neural Network using Sequential model from Keras API integrating TF

#Simple tf.keras Sequential Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

*Checking the number of GPUs*

```python
physical_devices = tf.config.list_physical_devices('GPU')
print("Num GPUs available: ", len(physical_devices))
```

*Sequential model*

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'), #Thus is actually the second layer so 1 has been passed as a parameter
    Dense(units=32, activation='relu'), #Most common activation
    Dense(units=2, activation='softmax') #Where we might find its outputs
])
```

```python
model.summary()
```

#Train our Neural Network based on the dataset we created and processed

*After building our model we are going to call model.compile function and it prepares model or training*

*That means getting everything in order that's needed before we actaully train the model*

*So this cell specifies which optimizer you are going to use, the type of loss, and metrics like 'accuracy'*

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

*So the model has been compiled and ready for training by now*

*And train occur when we call the following function:*

*By specifying the first parameter as x(our data input) stored in x_scaler_train_samples, y(our Target)*

*batch_size is how many samples are included during the process at once at one single neuron*

*Epochs is how many times your odel is gonna process, "do the math"*

*Shuffle by default is arealdy true, you don't have to specify de paramater*

*Verbose allows us to see the "output", and 2 is the highest value for it, so we get the highest level of the output*

*Check also by adding "validation_split" how well the models can generalize*

```python
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

#Confusion Matrix

*Importing the packs needed*

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

*Plot Confusion Matrix Function like (copied from tensorflow)*

```python
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion Matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[j, i],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effects'] #They are the corresponding labels for our test_data
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix') #So plot Paramater 1) Passing confusion matrix, classes for the confusion matrix, and a title to display it.
```
